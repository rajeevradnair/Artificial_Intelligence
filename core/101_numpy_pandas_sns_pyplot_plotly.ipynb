```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon',
                              'pastrami', 'honey ham', 'nova lox'],
                     'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
data['food']=data['food'].str.lower()
```

```python
def f1(k):
  dict={
  'bacon': 'piglet',
  'pulled pork': 'pig',
  'pastrami': 'cow',
  'corned beef': 'cow',
  'honey ham': 'pig',
  'nova lox': 'salmon'
  }
  return dict[k]

data['animal']=data['food'].map(f1)
data
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x=np.linspace(0,10,100)
plt.plot(x, np.sin(x), '_')
plt.plot(x, np.cos(x), '.')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))

plt.subplot(2,2,3)
plt.plot(x, np.tan(x))

plt.subplot(2,2,4)
plt.plot(x, 1/np.tan(x))
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('Volcano.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.hist(df['Country'])
plt.show()
```

```python
import plotly.express as px
fig=px.histogram(df, x='Country').update_xaxes(categoryorder='total descending')
fig.show()
```

```python
import plotly.graph_objects as go
labels=df['Status'].unique()
values=df['Status'].value_counts()
fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.show()
```

```python
fig = go.Figure(
    data=go.Scattergeo(
        lon=df['Longitude'],
        lat=df['Latitude']
    )
)
fig.update_layout(
    title='Volcanoes',
    geo_scope='world'
)
fig.show()
```

```python
type(fig)
```

```python
fig = go.Figure(
    data=go.Scattergeo(
        lon=df['Longitude'],
        lat=df['Latitude']
    )
)
fig.update_geos(
    projection_type="orthographic",
    landcolor="gold",
    oceancolor="MidnightBlue",
    showocean=True,
    lakecolor="LightBlue"
)
fig.show()
```
